```python
import pandas as pd
import numpy as np
import statsmodels.tools.tools as stattools
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
import os
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
adult = fetch_ucirepo(id=2) 
  
# data (as pandas dataframes) 
data = adult.data.features 
target = adult.data.targets

```

```python
# For reference
adult.variables
```

```python
# we only use Martial status and cap_gain and cap_lose
data = data[['marital-status','capital-gain','capital-loss']]
data
```

```python
# Make new variable called Cap_gain_losses
df_train, df_test, target_train, target_test = train_test_split(data,target)


df_train = df_train.copy()
df_train['Cap_gain_loss']= df_train['capital-gain'] + df_train['capital-loss'].abs()
df_train = df_train[['marital-status','Cap_gain_loss']]
```

```python
mar_df = pd.get_dummies(df_train['marital-status'],dummy_na=False)
mar_df = mar_df.astype(int)
mar_df
```

```python
X = pd.concat((df_train[['Cap_gain_loss']],mar_df),axis=1)
X
```

```python
outfile = os.getcwd() + "/cart01.png"
cart01 = DecisionTreeClassifier(criterion="gini",max_leaf_nodes = 5).fit(X,target_train)
export_graphviz(cart01,out_file=outfile)

os.system('ls')
```
